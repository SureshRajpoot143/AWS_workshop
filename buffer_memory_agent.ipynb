```python
from datetime import datetime

FILE_NAME = "user_memory.txt"

print("📝 User Memory Saver")
print("Type 'exit' to stop\n")

while True:

    user = input("You: ")

    if user.lower() == "exit":
        break

    timestamp = datetime.now().strftime(
        "%Y-%m-%d %H:%M:%S"
    )

    with open(
        FILE_NAME,
        "a",
        encoding="utf-8"
    ) as f:

        f.write(
            f"[{timestamp}] {user}\n"
        )

    print("✅ Saved")
```
