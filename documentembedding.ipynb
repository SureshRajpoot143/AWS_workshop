```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
doc1 = """
Python Developer with 3 years experience in
Machine Learning, SQL, AWS and Data Analysis.
"""

doc2 = """
Java Developer with Spring Boot,
MySQL and Microservices experience.
"""
```

```python
embedding1 = model.encode(doc1)
embedding2 = model.encode(doc2)

print(embedding1.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(
    [embedding1],
    [embedding2]
)

print(similarity[0][0])
```

```python
job_description = """
Looking for a Python Developer with SQL,
AWS and Machine Learning skills.
"""
```

```python
 resume = """
Experienced Python Developer.
Worked on AWS cloud projects,
SQL databases and Machine Learning models.
"""
```

```python
job_emb = model.encode(job_description)
resume_emb = model.encode(resume)

score = cosine_similarity(
    [job_emb],
    [resume_emb]
)[0][0]

print(f"Match Score: {score*100:.2f}%")
```
