```python
from datetime import datetime

SAFE_TIME = "09:00"

def attendance_agent(employee_name, check_in_time):
    safe_time = datetime.strptime(SAFE_TIME, "%H:%M")
    arrival_time = datetime.strptime(check_in_time, "%H:%M")

    if arrival_time <= safe_time:
        status = "SAFE ZONE"
    else:
        status = "DANGER ZONE"

    return {
        "Employee": employee_name,
        "Arrival Time": check_in_time,
        "Status": status
    }

name = input("Enter employee name: ")
time = input("Enter arrival time (HH:MM): ")

result = attendance_agent(name, time)

print("\nAttendance Report")
print("Employee:", result["Employee"])
print("Arrival Time:", result["Arrival Time"])
print("Status:", result["Status"])
```
