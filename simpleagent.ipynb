```python
from datetime import datetime

SAFE_TIME = "09:00"

def attendance_agent(employee_name, check_in_time):
    safe_time = datetime.strptime(SAFE_TIME, "%H:%M")
    arrival_time = datetime.strptime(check_in_time, "%H:%M")

    if arrival_time <= safe_time:
        status = "SAFE ZONE ✅"
    else:
        status = "DANGER ZONE ⚠️"

    return {
        "Employee": employee_name,
        "Arrival Time": check_in_time,
        "Status": status
    }

# Example
print(attendance_agent("Rahul", "08:55"))
print(attendance_agent("Amit", "09:15"))
```

```python
employees = [
    ("Rahul", "08:55"),
    ("Amit", "09:15"),
    ("Neha", "08:45"),
    ("Priya", "09:05")
]

for name, time in employees:
    print(attendance_agent(name, time))
```
