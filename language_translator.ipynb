```python
from deep_translator import GoogleTranslator

print("🌍 Language Translator")

while True:
    text = input("\nEnter text: ")

    if text.lower() == "exit":
        break

    target_lang = input("Translate to (hi, en, fr, de, es): ")

    translated = GoogleTranslator(
        source='auto',
        target=target_lang
    ).translate(text)

    print("\nTranslated Text:")
    print(translated)
```
